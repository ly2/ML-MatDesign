```python
def start_requests(url):
    print('getting', url) 
    return requests.get(url)

    

def get_info(rep):
    data = rep.json()
    for d in data['records']:
        yield {
            'title': d['title'],
            'abstract':d['abstract']
        }
        
    

def get_all_springer_data(year):
    baseurl = 'http://api.springernature.com/metadata/json?q=subject:{} year: {}&s={}&p=50&api_key='
    subjects = ['%22Materials%20Science%22', 'Chemistry', 'Physics']
    appcode = '9be60cd8ac2dc02001f490ea60cc7d17'
    appcode2 = 'c67d89b0dfd30a44db23368176e60ddf'
    appcode3 = 'c56d43d450736d376c21a7e0d530d32e'
    appcode4 = 'f0039a389720773fee9c6a9defb7eb5c'
    appcode5 = '94f901d3e83a66e0372c678176ef6a6a'
    appcode6 = '2954959a7e97ed368087f090bf2916d5'
    appcode7 = '600fee63e8f317659f15ee8a86733e9f'
    appcode10 = '6a28177f8d741f0212d1c1e3872f7bea'
    
    for subject in subjects:
        total_num = int(requests.get(baseurl.format(subject,year,'0')+appcode2).json()['result'][0]['total'])
        p = 1
        while p < total_num:
            r = start_requests(baseurl.format(subject,year,str(p))+appcode2)
            p += 50
            yield from get_info(r)
                
```

```python
years = list(range(1922,2019))

years.extend([1985,1990,1995, 2000])
for year in years:
    data = list(get_all_springer_data(year))
    s = json.dumps(data, ensure_ascii=False, indent=4)
    with open('data/springer{}.json'.format(year), 'w', encoding='utf-8') as f:
        f.write(s)
```

```python
ELE_ELENAMES = ELE + ELE_NAMES + [en.capitalize() for en in ELE_NAMES]
ELENAMES_Cap = ELE_NAMES + [en.capitalize() for en in ELE_NAMES]

# Elemement with the valence state in parenthesis.
ELEMENT_VALENCE_IN_PAR = regex.compile(r"^("+r"|".join(ELE_ELENAMES) + r")(\(([IV|iv]|[Vv]?[Ii]{0,3})\))$")
ELEMENT_DIRECTION_IN_PAR = regex.compile(r"^(" + r"|".join(ELE_ELENAMES) + r")(\(\d\d\d\d?\))")

# Exactly IV, VI or has 2 consecutive II, or roman in parenthesis: is not a simple formula.
VALENCE_INFO = regex.compile(r"(II+|^IV$|^VI$|\(IV\)|\(V?I{0,3}\))")

Num = regex.compile(r"^[+-]?\d*\.?\d+\(?\d*\)?+$", regex.DOTALL)
Num_UNIT = regex.compile(r"^([+-]?\d*\.?\d+\(?\d*\)?+)([\p{script=Latin}|Ω|μ]+.*)", regex.DOTALL)
PUNCT = list(string.punctuation) + ["\"", "“", "”", "≥", "≤", "×"]
```

```python
def save_obj(obj, name):
    with open(name + ".pkl", "wb") as f:
        pickle.dump(obj, f, pickle.HIGHEST_PROTOCOL)

def load_obj(name):
    with open(name + ".pkl", "rb") as f:
        return pickle.load(f)
```

```python
def tokenise(text, split_oxidation=True, keep_sent=True):
    def split_token(token, so=split_oxidation):
        elem_with_valence = ELEMENT_VALENCE_IN_PAR.match(token) if so else None
        nr_unit = Num_UNIT.match(token)
        if nr_unit is not None and nr_unit.group(2) in SPLIT_UNITS:
            return [nr_unit.group(1), nr_unit.group(2)]
        elif elem_with_valence is not None:
            return [elem_with_valence.group(1), elem_with_valence.group(2)]
        else:
            return [token]
        
    para = Paragraph(text)
    tokens = para.tokens
    toks = []
    for sentence in tokens:
        if keep_sent:
            toks.append([])
            for tok in sentence:
                toks[-1] += split_token(tok.text, so=split_oxidation)
        else:
            for tok in sentence:
                toks += split_token(tok.text, so=split_oxidation)
    return toks
```

```python
def make_phrases(phraser, sentence, reps=2):
    while reps > 0:
        sentence = phraser[sentence]
        reps -= 1
    return sentence

def is_number(s):
    if s is 'Inf':
        return True
    return Num.match(s.replace(",", "")) is not None

def is_element(txt):
    try:
        Element(txt)
        return True
    except ValueError:
        return False

def is_simple_formula(text):
    if VALENCE_INFO.search(text) is not None:
            # 2 consecutive II, IV or VI should not be parsed as formula.
            # Related to valence state, so don"t want to mix with I and V elements.
        return False
    elif any(char.isdigit() or char.islower() for char in text):
        try:
            if text in ["O2", "N2", "Cl2", "F2", "H2"]:
                return True
            composition = Composition(text)
                # Has to contain more than one element, single elements are handled differently.
            if len(composition.keys()) < 2 or any([not is_element(key) for key in composition.keys()]):
                return False
            return True
        except (CompositionError, ValueError):
            return False
        except:
            print(text)
            return False
    else:
        return False

def normalized_formula(formula, max_denominator=1000):
    try:
        formula_dict = Composition(formula).get_el_amt_dict()
        return get_ordered_integer_formula(formula_dict, max_denominator)
    except (CompositionError, ValueError):
        return formula
    
def get_ordered_integer_formula(el_amt, max_denominator=1000):
    g = gcd_float(list(el_amt.values()), 1 / max_denominator)
    d = {k: round(v / g) for k, v in el_amt.items()}
    formula = ""
    for k in sorted(d):
        if d[k] > 1:
            formula += k + str(d[k])
        elif d[k] != 0:
            formula += k
    return formula


def remove_accent(txt):
    return unidecode.unidecode(txt) if len(txt) > 1 else txt
```

```python
elem_name_dict = {en: es for en, es in zip(ELE_NAMES, ELE)}
def process(tokens, exclude_punct=False, convert_num=True, normalize_materials=True, remove_accents=True, make_phrases=False, split_oxidation=True):
    processed, mat_list = [], []

    for i, tok in enumerate(tokens):
        if exclude_punct and tok in PUNCT:  # Punctuation.
            continue
        elif convert_num and is_number(tok):
            try:
                if tokens[i - 1] == "(" and tokens[i + 1] == ")" or tokens[i - 1] == "〈" and tokens[i + 1] == "〉":
                    pass
                else:
                    tok = "<nUm>"
            except IndexError:
                tok = "<nUm>"
        elif tok in ELENAMES_Cap:  # Chemical element name.
                # Add as a material mention.
            mat_list.append((tok, elem_name_dict[tok.lower()]))
            tok = tok.lower()
        elif is_simple_formula(tok):  # Simple chemical formula.
            normalized_f = normalized_formula(tok)
            mat_list.append((tok, normalized_f))
            if normalize_materials:
                tok = normalized_f
        elif (len(tok) == 1 or (len(tok) > 1 and tok[0].isupper() and tok[1:].islower())) and tok not in ELE and tok not in SPLIT_UNITS and ELEMENT_DIRECTION_IN_PAR.match(tok) is None:
                # To lowercase if only first letter is uppercase (chemical elements already covered above).
            tok = tok.lower()

        if remove_accents:
            tok = remove_accent(tok)

        processed.append(tok)

    if make_phrases:
        processed = make_phrases(processed, reps=2)

    return processed, mat_list
```

```python
def exclude_words(phrasegrams, words):
    """Given a list of words, excludes those from the keys of the phrase dictionary."""
    new_phrasergrams = {}
    words_re_list = []
    for word in words:
        we = regex.escape(word)
        words_re_list.append("^" + we + "$|^" + we + "_|_" + we + "$|_" + we + "_")
    word_reg = regex.compile(r""+"|".join(words_re_list))

    for gram in tqdm(phrasegrams):
        valid = True
        for sub_gram in gram:
            if word_reg.search(sub_gram.decode("unicode_escape", "ignore")) is not None:
                valid = False
                break
            if not valid:
                continue
        if valid:
            new_phrasergrams[gram] = phrasegrams[gram]
    return new_phrasergrams

def wordgrams(sent, depth, pc, th, ct, et, d=0):
    if depth == 0:
        return sent, None
    else:
        """Builds word grams according to the specification."""
        phrases = Phrases(
            sent,
            common_terms=ct,
            min_count=pc,
            threshold=th)

        grams = Phraser(phrases)
        grams.phrasegrams = exclude_words(grams.phrasegrams, et)
        d += 1
        if d < depth:
            return wordgrams(grams[sent], depth, pc, th, ct, et, d)
        else:
            return grams[sent], grams
        
def keep_simple_formula(word, count, min_count):
    if is_simple_formula(word):
        return gensim.utils.RULE_KEEP
    else:
        return gensim.utils.RULE_DEFAULT
```

```python
def compute_epoch_accuracies(root, prefix, analogy_file):
    filenames = glob.glob(os.path.join(root, prefix+"_epoch*.model"))
    nr_epochs = len(filenames)
    accuracies = dict()
    losses = [0] * nr_epochs
    for filename in filenames:
        epoch = int(re.search("\d+\.model", filename).group()[:-6])
        m = Word2Vec.load(filename)
        losses[epoch] = m.get_latest_training_loss()
        sections = m.wv.accuracy(analogy_file)
        for sec in sections:
            if sec["section"] not in accuracies:
                accuracies[sec["section"]] = [0] * nr_epochs
            correct, incorrect = len(sec["correct"]), len(sec["incorrect"])
            if incorrect > 0:
                accuracy = correct / (correct + incorrect)
            else:
                accuracy = 0
            accuracies[sec["section"]][epoch] = (correct, incorrect, accuracy)
        save_obj(accuracies, os.path.join("models", prefix + "_accuracies"))
        save_obj(np.concatenate([np.array([losses[0]]), np.diff(losses)]), os.path.join("models", prefix + "_loss"))

class EpochSaver(CallbackAny2Vec):
    def __init__(self, path_prefix):
        self.path_prefix = path_prefix
        self.epoch = 0

    def on_epoch_end(self, m):
        output_path = "{}_epoch{}.model".format(self.path_prefix, self.epoch)
        print("Save model to {}.".format(output_path))
        m.save("tmp/"+output_path)
        self.epoch += 1
```
